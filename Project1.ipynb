```python
#find Branch wise Payment methond  distribution in all branches
plt.figure(figsize=(14,6))
plt.style.use('classic')
ax=sns.countplot(x="Payment",hue="Branch",data=df,palette="tab20")
ax.set_title(label="Payment distribution in all branches",fontsize=25)
ax.set_xlabel(xlabel="payment method",fontsize=16)
ax.set_ylabel(ylabel="People count",fontsize=16)
```

```python
#Does gross income affect customer rating
sns.set_style('darkgrid')
sns.scatterplot(x=df['Rating'],y=df['gross income'])
plt.title("Gross income Vs Rating")
```

```python
#find rating distribution beteween branches
plt.figure(figsize=(8,4))
ax=sns.boxplot(x="Branch",y="Rating",data=df,palette="RdYlBu")
ax.set_title("Rating distribution between branches",fontsize=25)
ax.set_xlabel(xlabel="Branches",fontsize=16)
ax.set_ylabel(ylabel="Rating distribution",fontsize=16)
plt.grid()
```

```python
#Can you see any time trend in gross income?
df1=df.select_dtypes(include='number')
df1.groupby(df.index).mean()
```

```python
sns.lineplot(x=df1.groupby(df.index).mean().index,
            y=df1.groupby(df.index).mean()['gross income'])
```

```python
#which product line help you to genrate the most income

cat=df[["Product line","gross income"]].groupby(['Product line']).sum().reset_index()
plt.figure(figsize=(20,8))
sns.barplot(x='Product line',y='gross income',data=cat)
plt.title("Group income VS Product line")
```

```python
#What is the spending pattern of both males and fenaless ,meaning in which category do they spend more?
plt.figure(figsize=(16,6))
plt.title('Total Monthly transaction by "Gender')
sns.countplot(x=df['Product line'],hue=df.Gender,palette='Blues')
```

```python
#How MAny Products bought by customer?
plt.figure(figsize=(12,6))
sns.distplot(x=df['Quantity'])
```

```python
#Which day of the Week has maximum sales? Datatype casting
df['Date']=pd.to_datetime(df['Date'])
df['weekday']=df['Date'].dt.day_name()
df.set_index('Date',inplace=True)
df.head()
```

```python
df.info()
```

```python
plt.figure(figsize=(8,6))
plt.title('Daily Sales by Day of the week')
sns.countplot(x=df['weekday'])
```

```python
#which hour of the day is busiest
df['Time']=pd.to_datetime(df['Time'])
df['Hour']=(df['Time']).dt.hour
df['Hour'].unique()
```

```python
sns.lineplot(x="Hour",y='Quantity',data=df).set_title("Product Sales per Hour")
```

```python
#ehich product line should your supermarket focus on
plt.figure(figsize=(12,6))
sns.barplot(y=df['Product line'],x=df['Rating'])
```
